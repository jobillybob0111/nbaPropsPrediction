# MVP Training Data Summary

Loads `exports/nba_training_mvp_v1.csv` and prints summary statistics.
This notebook **does not** modify or write any data.

```python
from pathlib import Path
import numpy as np
import pandas as pd

ROOT = Path.cwd()
DATA_PATH = ROOT / "exports" / "nba_training_mvp_v1.csv"

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
df.dtypes
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

df[numeric_cols].describe().T
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
df.select_dtypes(exclude=[np.number]).nunique().sort_values(ascending=False)
```
